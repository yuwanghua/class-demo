```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.style.use('ggplot')
```

## 读取数据

```python
df1 = pd.read_csv('moves_new.csv')
df2 = pd.read_csv('moves_new1.csv')
```

## 合并数据

```python
df = pd.concat([df1,df2],axis=0,ignore_index=True)
df
```

## 数据清洗+数据预处理

```python
df = df.drop(['country','language'],axis=1)
```

```python
df.info()
```

```python
df['1星'] = df['1星'].apply(lambda x:x.strip('%')).astype("float")
df['2星'] = df['2星'].apply(lambda x:x.strip('%')).astype("float")
df['3星'] = df['3星'].apply(lambda x:x.strip('%')).astype("float")
df['4星'] = df['4星'].apply(lambda x:x.strip('%')).astype("float")
df['5星'] = df['5星'].apply(lambda x:x.strip('%')).astype("float")
```

```python
import re
df['comment'] = df['comment'].astype('int')
df['score'] = df['score'].astype('float')
df['time'] = df['time'].apply(lambda x:x.strip("分钟,分钟(剧场版),分钟(中国大陆),分钟(印度),分钟(香港),分钟(3D重映),分钟(导演剪辑)\
                                               ,分钟(泰),分钟(国际),分钟(韩)，分钟(美),分钟(欧洲),分钟(日本),分钟(数字),USA: \
                                               ,分钟(公),Japan:,分钟(中国大陆/),分钟(台湾),分钟(圣丹斯电影节),分钟(德),分钟(戛纳)\
                                               ")).astype('int')
```

```python
df
```

```python
import re
def transformCountry(d):
    list1 = re.findall(r'\(.*?\)',d) 
    if len(list1)>0:
        return list1[0]
    else:
        return np.nan
df['country'] = df['date'].map(transformCountry)
df
```

```python
# 看一下为什么正则表达式取不出国家
df[df['country'].isnull()]
```

```python
def transformDate(d):
    return d[2:13].split('(')[0]
df['date'] = df['date'].map(transformDate)
df
```

### 根据电影type分类关于score的密度图

```python
types = df['type'].value_counts() 
types_count = types[types>9].index
plt.subplots(figsize=(12, 10))

# Plot each building
for type in types_count:
    # Select the building type
    subset = df[df['type'] == type]
    data_1=subset['score'].dropna()
    data_1.plot(kind="density",label = type, alpha = 0.8);
    
plt.xlabel('score', size = 20)
plt.ylabel('density', size = 20)
plt.title('电影类型的density图', size = 28)
plt.legend()
```

### 根据上映国家分类score的密度函数图

```python
countrys = df['country'].value_counts() 
countrys_count = countrys[countrys>40].index
plt.subplots(figsize=(12, 10))

# Plot each building
for country in countrys_count:
    # Select the building type
    subset = df[df['country'] == country]
    data_1=subset['score'].dropna()
    data_1.plot(kind="density",label = country, alpha = 0.8);
    
plt.xlabel('score', size = 20)
plt.ylabel('density', size = 20)
plt.title('电影上映国家的density图', size = 28)
plt.legend()
```
